```python
import pandas as pd

# Read the CSV file into a DataFrame
file_path = '/content/FLB_ARIES_data.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Remove columns with all blank (NaN) values
df = df.dropna(axis=1, how='all')

#--------------------remove duplicate value---------------------------------------------------------
# Create the new 'Final_Entity' column as some lot was move in/out using different tool --> messy data, need to dig separately
df['final_entity'] = df.apply(lambda row: row['entity'] if row['last_entity'] == row['entity'] else 'mix_FLB201/202', axis=1)
# Remove the 'Entity' column if it exists
if 'entity' in df.columns:
    df = df.drop(columns=['entity'])
# Remove the 'Entity' column if it exists
if 'last_entity' in df.columns:
    df = df.drop(columns=['last_entity'])
# Remove duplicate rows
df = df.drop_duplicates()

# Identify duplicates in 'sub_mark_id_1'
duplicates = df[df.duplicated('sub_mark_id_1', keep=False)]

# Filter to keep only rows where 'final_entity' is 'mix_FLB201/202'
to_keep = duplicates[duplicates['final_entity'] == 'mix_FLB201/202']

# Remove all duplicates from the original DataFrame
df = df.drop_duplicates('sub_mark_id_1', keep=False)

# Append the rows to keep back to the DataFrame
df = pd.concat([df, to_keep]).sort_index()

#---------------complete remove duplicaton-------------------------

# Display the cleaned DataFrame
#print(df_cleaned)
# Optionally, save the cleaned DataFrame to a new CSV file
df.to_csv('/content/FLB_ARIES_data_filter.csv', index=False)
```
